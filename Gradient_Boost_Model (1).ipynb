Loading Panda

```python
import pandas as P
```

Opening Data

```python
D = P.read_csv("Nigeria Crash Data U.csv")
```

Columns

```python
print(D.columns)
```

X = Input Features , Y = Output Features

```python
X = D.drop (["Num_Killed"] , axis = 1)
Y = D["Num_Killed"]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

Applying Model (Gradient Boosting Regressor is a model that predicts numbers (continuous values) by adding small trees step by step, reducing the error)

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
GBR = GradientBoostingRegressor (random_state = 42)
```

```python
GBR.fit(X_Train , Y_Train)
```

Prediction

```python
Y_P = GBR.predict(X_Test)
```

Performance

```python
import numpy as N
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error
```

```python
R2 = r2_score(Y_Test , Y_P)
MSE = mean_squared_error(Y_Test , Y_P)
RMSE = N.sqrt(MSE)
MAE = mean_absolute_error(Y_Test , Y_P)
```

```python
print("R2 Score : " , R2 * 100 , "%")
print("\n")
print("MSE : " , MSE)
print("\n")
print("RMSE : " , RMSE)
print("\n")
print("MAE : " , MAE)
```

```python
Train_Score = GBR.score(X_Train , Y_Train)
print("Train Score : " , Train_Score * 100 , "%")
print("\n")
Test_Score = GBR.score(X_Test , Y_Test)
print("Test Score : " , Test_Score * 100 , "%")
```
