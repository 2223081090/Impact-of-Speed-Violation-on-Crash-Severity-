Loading Python Library Panda

```python
import pandas as P
```

Opening Dataset in Python

```python
D = P.read_csv("Nigeria Crash Data U.csv")
```

Printing Columns Of Datasets

```python
print(D.columns)
```

Giving X = Input Features (Column) , Y = Output Features (Column)

```python
X = D.drop("Num_Killed" , axis = 1)
Y = D["Num_Killed"]
```

Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

Printing Train Test Splits shape Variables

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

Applyting Ridge Regression with hyper parameter tuning (Ridge Regression = Linear Regression but reduces overfitting)

Set Alpha Value



```python
AV = [0.001 , 0.01 , 0.1 , 1 , 10 , 100]
```

Apply Model

```python
from sklearn.linear_model import Ridge
```

```python
R = Ridge ()
```

Apply hyperparameter on Model and train

```python
from sklearn.model_selection import GridSearchCV
```

```python
PG = {
   "alpha" : AV
}
```

```python
G = GridSearchCV(R , PG , cv = 5 , scoring = "r2")
```

```python
G.fit(X_Train , Y_Train)
```

Training with best alpha not overfitting

```python
BA = G.best_params_["alpha"]
```

```python
RB = Ridge(alpha = BA)
```

```python
RB.fit(X_Train , Y_Train)
```

Prediction of Model

```python
YP = RB.predict(X_Test)
```

Checking Model Performance

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
R2 = RB.score(X_Test , Y_Test)
```

```python
print("R2 Score : " , R2 * 100 , "%")
```

```python
MSE = mean_squared_error(Y_Test , YP)
print(MSE)
```

```python
import numpy as N
RMSE = N.sqrt(MSE)
print(RMSE)
```

```python
MAE = mean_absolute_error(Y_Test , YP)
print(MAE)
```

```python
Train_Score = RB.score(X_Train , Y_Train)
print(Train_Score * 100 , "%")
print("\n")
Test_Score = RB.score(X_Test , Y_Test)
print(Test_Score * 100 , "%")
```
