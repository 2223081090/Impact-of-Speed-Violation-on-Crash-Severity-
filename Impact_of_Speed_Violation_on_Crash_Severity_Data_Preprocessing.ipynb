```python
D["Other_Factors"] = D["Other_Factors"].abs()
```

```python
Q1 = D.quantile(0.25)
Q3 = D.quantile(0.75)
IQR = Q3 - Q1
```

```python
Outlier = (D < (Q1 - 1.5 * IQR)) | (D > (Q3 + 1.5 * IQR))
print(Outlier)
print(Outlier.sum())
```

```python
L = Q1 - 1.5 * IQR
U = Q3 + 1.5 * IQR
D = D.clip(lower = L, upper = U, axis=1)
```

```python
print(D.columns)
```

```python
Columns_Drop = D["Other_Factors"]
```

```python
D.drop(columns = ["Other_Factors"] , inplace = True)
```
