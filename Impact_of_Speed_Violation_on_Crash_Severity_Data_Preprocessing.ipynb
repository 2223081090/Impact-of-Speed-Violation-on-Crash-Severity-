Make "Quarter" and "State" Column in numeric value

```python
D["Quarter"] = P.to_numeric(D["Quarter"] , errors = "coerce")
```

```python
D["State"] = P.to_numeric(D["State"] , errors = "coerce")
```

Watch Again Data Types

```python
print(Check_Negetive)
```

Make "Other_Factors" Columns Value Positive

```python
D["Other_Factors"] = D["Other_Factors"].abs()
```

Check The Dataset Outlier

```python
Q1 = D.quantile(0.25)
Q3 = D.quantile(0.75)
IQR = Q3 - Q1
```

Removing Outlier

```python
Outlier = (D < (Q1 - 1.5 * IQR)) | (D > (Q3 + 1.5 * IQR))
print(Outlier)
print(Outlier.sum())
```

```python
L = Q1 - 1.5 * IQR
U = Q3 + 1.5 * IQR
D = D.clip(lower = L, upper = U, axis=1)
```

Checking Outlier Again

```python
print(L)
print("\n")
print(U)
print("\n")
print(D)
```
