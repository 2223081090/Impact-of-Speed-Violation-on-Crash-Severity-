Using Library Panda of Python

```python
import pandas as P
```

Opening Dataset in Python

```python
D = P.read_csv("Nigeria Crash Data U.csv")
```

Watching Columns

```python
print(D.columns)
```

Giving X = Input Features (Columns) , Y = Output Feature (Column)

```python
X = D.drop(["Num_Killed"] , axis = 1)
Y = D["Num_Killed"]
```

Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

Printing Variables

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

Applying Lasso Regression Model (Lasso = Linear Regression which completely eliminates some unnecessary things)

Training Model with hyper parameter tuning

```python
from sklearn.linear_model import Lasso
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
AV = [0.001 , 0.01 , 0.1, 1 , 10 , 100]
```

```python
M = Lasso ()
```

```python
PG = {"alpha" : AV}
```

```python
G = GridSearchCV(M , PG , cv = 5 , scoring = "r2")
```

```python
G.fit(X_Train , Y_Train)
```

Training Model with Best alpha fot not overfitting

```python
BA = G.best_params_["alpha"]
```

```python
LF = Lasso(alpha = BA)
```

```python
LF.fit(X_Train , Y_Train)
```

Prediction

```python
YP = LF.predict(X_Test)
```

Checking Model Performance

```python
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error
import numpy as N
```

```python
R2 = r2_score(Y_Test , YP)
```

```python
print(R2 * 100 , "%")
```

```python
MAE = mean_absolute_error(Y_Test , YP)
print(MAE)
```

```python
MSE = mean_squared_error(Y_Test , YP)
print(MSE)
```

```python
RMSE = N.sqrt(MSE)
print(RMSE)
```

```python
Train_Score = LF.score(X_Train , Y_Train)
print(Train_Score * 100 , "%")
print("\n")
Test_Score = LF.score(X_Test , Y_Test)
print(Test_Score * 100 , "%")
```
