Loading Python Library

```python
import pandas as P
```

Open Dataset in python

```python
D = P.read_csv("Nigeria Crash Data U.csv")
```

Printing all columns of Dataset

```python
print(D.columns)
```

Setting X = Input Features (Columns) and Y = Output Feature (Column)

```python
X = D.drop(columns=["Num_Killed"] , axis = 1)
Y = D["Num_Killed"]
```

Applying Train Test Split to Dataset . 80% percent training and 20% testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

Printing Shape of X_Train , Y_Train , X_Test , Y_Test

```python
print(X_Train.shape , X_Test.shape)
print("\n")
print(Y_Train.shape , Y_Test.shape)
```

Applying Linear Regression Model to Nigeria Crash Data Dataset New (Linear Regression means predicting new values ​​by looking at old data)

Applying Model and training


```python
from sklearn.linear_model import LinearRegression
```

```python
L = LinearRegression()
```

```python
L.fit(X_Train , Y_Train)
```

Prediction

```python
Y_P = L.predict(X_Test)
```

Performance

```python
import numpy as N
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
R2 = r2_score(Y_Test , Y_P)
print(R2 * 100 , "%")
```

```python
MAE = mean_absolute_error(Y_Test , Y_P)
print(MAE)
```

```python
MSE = mean_squared_error(Y_Test , Y_P)
print(MSE)
```

```python
RMSE = N.sqrt(MSE)
print(RMSE)
```

```python
Train_Score = L.score(X_Train , Y_Train)
print(Train_Score * 100 , "%")
print("\n")
Test_Score = L.score(X_Test , Y_Test)
print(Test_Score * 100 , "%")
```
